```python
data_nbc = pd.read_csv("./gender_classification.csv")
data_nbc
```

```python
x_data = data_nbc.drop(["gender"],axis = 1)
y_data = data_nbc.gender.values
```

```python
x_data
```

```python
y_data
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.3, random_state=1)
```

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(x_train, y_train)

print("print Train for accuracy of NBC algo: ", gnb.score(x_train,y_train))
print("print Test for accuracy of NBC algo: ", gnb.score(x_test,y_test))
```

Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()

mnb.fit(x_train, y_train)

print("print Train for accuracy of NBC algo: ", mnb.score(x_train,y_train))
print("print Test for accuracy of NBC algo: ", mnb.score(x_test,y_test))
```

Bernouli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()

bnb.fit(x_train, y_train)

print("print Train for accuracy of NBC algo: ", bnb.score(x_train,y_train))
print("print Test for accuracy of NBC algo: ", bnb.score(x_test,y_test))
```

```python
import pandas as pd

def create_frequency_table(data, features, target):
    """
    Creates a frequency table for a given dataset.

    Args:
        data: A Pandas DataFrame containing the dataset.
        features: A list of feature names.
        target: The name of the target variable (optional).

    Returns:
        A Pandas DataFrame representing the frequency table.

    Raises:
        ValueError: If the target variable is not specified for categorical features.
    """

    # Handle categorical features
    if data[features].dtypes.all() == object:
        if target is None:
            raise ValueError("Target variable is required for categorical features.")
        table = pd.crosstab(data[features], data[target])
        table.rename_axis(['Feature Value'], axis=1, inplace=True)
        table.rename_axis('Class', axis=0, inplace=True)
    # Handle numerical features (assuming counts or frequencies)
    else:
        table = data[features].describe(include='all')
        table = table.unstack().fillna(0)  # Fill missing values with 0s

    return table
```

```python
def calculate_likelihood(frequency_table, feature, feature_value, target_class):
    """
    Calculates the likelihood of a feature-value combination for a specific class.

    Args:
        frequency_table: A Pandas DataFrame representing the frequency table.
        feature: The feature name.
        feature_value: The value of the feature.
        target_class: The target class.

    Returns:
        The likelihood of the feature-value combination.
    """

    count = frequency_table.loc[target_class, feature_value]
    total = frequency_table.loc[target_class].sum()
    return count / total if total > 0 else 0
```

```python
features=['long_hair','forehead_width_cm','forehead_height_cm','nose_wide','nose_long','lips_thin','distance_nose_to_lip_long']
```

```python
frequency_table = create_frequency_table(data, features, 'gender')
print(frequency_table)
```

```python
frequency_table = calculate_likelihood(frequency_table, data, features, 'gender')
print(frequency_table)
```
